```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from skimage import io
```

```python
SEED = 12345
np.random.seed(SEED)

path = 'C:/Users/Gavela Maculuve/PycharmProjects/Ritcher/CS/Pneumonia2'
trainpath = path + '/train'
testpath = path + '/test'
validpath = path + '/valid'
classes = ['BACTERIAL', 'NORMAL', 'VIRAL']

NUMCLASSES = len(classes)
VALIDATION = 0.2
```

```python
for c in classes:
    filelist = os.listdir(trainpath + '/' + c)
    print(c,len(filelist))
```

```python
for c in classes:
    print(c)
    filelist = os.listdir(trainpath + '/' + c)
    if c == 'NORMAL':
        try:
            filelist.remove('.DS_Store')
        except:
            print(".DS_Store file not found")
        np.random.shuffle(filelist)
        valid = np.int16(VALIDATION * len(filelist))
        validfiles = filelist[:valid]
        trainfiles = filelist[valid:]
    else:
        persons = {}
        try:
            filelist.remove('.DS_Store')
        except:
            print(".DS_Store file not found")
        for file in filelist:
            if file.startswith('person'):
                person = (file.split('_'))[0]
                if person in persons:
                    persons[person] += 1
                else:
                    persons[person] = 1
        index = np.array(range(len(persons)))
        np.random.shuffle(index)
        valid_limit = np.int16(VALIDATION * len(filelist))
        
        validpersons = []
        validcounts = 0
        person_list = np.array(list(persons.keys()))
        for p in person_list[index]:
            if validcounts < valid_limit:
                validpersons.append(p)
                validcounts += persons[p]
        # trainpersons = persons[valid:]
        trainfiles = []
        validfiles = []
        
        for file in filelist:
            person = (file.split('_'))[0]
            if person in validpersons:
                validfiles.append(file)
            else:
                trainfiles.append(file)
    trainfiles = np.array(trainfiles)
    validfiles = np.array(validfiles)
    print(trainfiles.shape)
    print(validfiles.shape)

    for i in range(len(validfiles)):
        img = io.imread(trainpath+'/' + c + "/" + validfiles[i])
        print("Saving: ", trainpath+'/' + c + "/" + validfiles[i], 'to', validpath+'/' + c + "/" + validfiles[i])
        io.imsave(validpath+'/' + c + "/" + validfiles[i], img)
    for i in range(len(trainfiles)):
        img = io.imread(trainpath+'/' + c + "/" + trainfiles[i])
        print("Saving: ", trainpath+'/' + c + "/" + trainfiles[i], 'to', trainpath+'/' + c + "1/" + trainfiles[i])
        io.imsave(trainpath+'/' + c + "1/" + trainfiles[i], img)
```
